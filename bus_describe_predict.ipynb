```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
train.info()
```

```python
train['date'] = pd.to_datetime(train['date'])
train['weekday'] = train['date'].dt.weekday
train = pd.get_dummies(train,columns=['weekday'])
```

```python
test['date'] = pd.to_datetime(test['date'])
test['weekday'] = test['date'].dt.weekday
test = pd.get_dummies(test,columns=['weekday'])
```

```python
train['in_out'].value_counts()
```

```python
train['in_out'] = train['in_out'].map({'시내':0,'시외':1})
test['in_out'] = test['in_out'].map({'시내':0,'시외':1})
```

```python
import geopy.distance
```

```python
coords_jejusi = (33.500770, 126.522761) #제주시의 위도 경도
coords_seoquipo = (33.259429, 126.558217) #서귀포시의 위도 경도


train['dis_jejusi'] = [geopy.distance.geodesic((train['latitude'].iloc[i],train['longitude'].iloc[i]), coords_jejusi).km for i in range(len(train))]
train['dis_seoquipo'] = [geopy.distance.geodesic((train['latitude'].iloc[i],train['longitude'].iloc[i]), coords_seoquipo).km for i in range(len(train))]
```

```python
train['dis_jejusi'].corr(train['18~20_ride'])
```

```python
train['dis_seoquipo'].corr(train['18~20_ride'])
```

```python
import seaborn as sns
c = train.corr()
sns.heatmap(c)
```

```python
test['dis_jejusi'] = [geopy.distance.geodesic((test['latitude'].iloc[i],test['longitude'].iloc[i]), coords_jejusi).km for i in range(len(test))]
test['dis_seoquipo'] = [geopy.distance.geodesic((test['latitude'].iloc[i],test['longitude'].iloc[i]), coords_seoquipo).km for i in range(len(test))]
```

```python
input_var=['in_out','latitude','longitude','6~7_ride', '7~8_ride', '8~9_ride', '9~10_ride',
       '10~11_ride', '11~12_ride', '6~7_takeoff', '7~8_takeoff', '8~9_takeoff',
       '9~10_takeoff', '10~11_takeoff', '11~12_takeoff','weekday_0', 'weekday_1', 'weekday_2', 'weekday_3', 'weekday_4',
       'weekday_5', 'weekday_6', 'dis_jejusi', 'dis_seoquipo']
target=['18~20_ride']
```

```python
X_train=train[input_var]
y_train=train[target]

X_test=test[input_var]
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=1217)
```

```python
rf.fit(X_train,y_train)
```

```python
test['18~20_ride'] = rf.predict(X_test)
test[['id','18~20_ride']].to_csv("subm_baseline.csv",index=False)
```
